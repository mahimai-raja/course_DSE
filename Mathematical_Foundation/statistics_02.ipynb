# Statistics - II

## Inferential statistics :

        Inferential Statistics is used to draw inferences beyond the immediate data available.        
<br>
        
## Sampling Distribution :

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A Sampling Distribution is a probability distribution of a statistic obtained through a large number of samples drawn from a specific population.
- Can be derived for sample mean, sample median, sample mode , etc
<br><br>

## Central limit theory :

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;It states that when plotting a sampling distribution of means, the mean of sample means will be equal to the population mean.
<br><br>
**Points to note :**

1. Central Limit Theorem holds true irrespective of the type of distribution of the population.
2. Now, we have a way to estimate the population mean by just making repeated observations of samples of a fixed size.
3. Greater the sample size, lower the standard error and greater the accuracy in determining the population mean from the sample mean.

## Confidence Interval :



&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The confidence interval is a type of interval estimate from the sampling distribution which gives a range of values in which the population statistic may lie.

# Hypothesis Testing

## Null Hypothesis 
    Null Hypothesis being the sample statistic to be equal to the population statistic.
<br>

## Alternate Hypothesis
    Alternate hypothesis states that the sample statistics is significantly different from the population.
**NOTE :**
- Most of the time our job is to disprove the null hypothesis
<br><br>


## T-Test :
    A t-test is a statistical test that is used to compare the means of two groups. 
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;It is often used in hypothesis testing to determine whether a process or treatment has an effect on the population of interest,
**NOTE :**
- You can also compare mean by : **ANOVA** (Analysis of Variance)
