## What is Statistics ?
            Statistics is the branch of mathematics dealing with the collection of data, analysis of data and   interpretation of data.

### Types of Statistics :
- Descriptive Statistics
- Inferential Statistics
<br><br>

### 1) Descriptive Statistics :
    Descriptive statistics is used to define or describe an entire dataset.
    
### Data can be described by:
1. Measure of Central Tendancy
2. Measure of Dispersion
3. Shape of the Data <br><br>

#### 1) Measure of Central Tendancy
        Measure of central tendancy is used to describe a entire dataset with a single value. The three most    common measure of central tendancy are mean, median and mode.
        
#### Mean 
    The “Mean” is the average of the data.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Average can be identified by summing up all the numbers and then dividing them by the number of observation.

    Mean = X1 + X2 + X3 +… +  Xn / n
    
#### Median

    Median is the 50%th percentile of the data. It is exactly the center point of the data.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Median can be identified by ordering the data and splits the data into two equal parts and find the number. It is the best way to find the center of the data.

#### Mode 

    Mode is frequently occurring data or elements.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;If an element occurs the highest number of times, it is the mode of that data. If no number in the data is repeated, then there is no mode for that data. There can be more than one mode in a dataset if two values have the same frequency and also the highest frequency.

    The mode can be calculated for both quantitative and qualitative data.
    NOTE : But Mean and Median only for numerical data.
<br>

#### 2) Measure of Dispersion
    The dispersion is the “Spread of the data”. It measures how far the data is spread.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In most of the dataset, the data values are closely located near the mean. On some other dataset, the values are widely spread out of the mean. These dispersions of data can be measured by 

1. Inter Quartile Range ( IQR )

2. Range

3. Standard Deviation

4. Variance

#### Inter Quartile Range :


**Outliers :** <br>
Outliers are extreme behaviours. An outlier is a data point that differs significantly from other observations. It can cause serious problems in analysis.<br><br>

Quartiles are special percentiles. <br>

1st Quartile Q1 is the same as the 25th percentile.<br>

2nd Quartile Q2 is the same as 50th percentile.<br>

3rd Quratile Q3 is same as 75th percentile<br>

Steps to find quartile and percentile<br>
- The data should sorted and ordered from the smallest to the largest.<br>

- For Quartiles, ordered data is divided into 4 equal parts.<br>

- For Percentiles, ordered data is divided into 100 equal parts.<br>

        Inter Quartile Range is the difference between the third quartile(Q3) and the first Quartile (Q1)

        IQR = Q3- Q1
        
#### Range :
The range is the difference between the largest and the smallest value in the data.

    Max – Min = Range
#### Standard Deviation :
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The most common measure of spread is the standard deviation.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The Standard deviation is the measure of how far the data deviates from the mean value.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The standard deviation formula varies for population and sample. Both formulas are similar, but not the same.

Symbol used for Sample Standard Deviation  –  “s” (lowercase)<br>
Symbol used for Population Standard Deviation – “σ” (sigma, lower case)
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Standard deviation $\sigma={\sqrt {\frac {\sum(x_{i}-{\mu})^{2}}{N}}}$
#### Z-Score :
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Z-score indicates how much a given value differs from the standard deviation.
<br><br>
#### Varience :
The variance is a measure of variability. It is the average squared deviation from the mean.

The symbol $σ^2$ represents the population variance and the symbol for s2 represents sample variance.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Varience $σ^2$ = $\frac{\sum (x_i - \bar{x})^2}{n - 1}$<br><br>


# 3. Shape of the Data

The shape of the data is important because making a decision about the probability of data is based on its shape.

The shape of the data can be measured by two methodologies.

1. Symmetric

2. Skewness

3. Kurtosis

### Symmetric
In the symmetric shape of the graph, the data is distributed the same on both sides.

In symmetric data, the mean and median are located close together.



## Skewness
Skewness is the measure of the asymmetry of the distribution of data.

The data is not symmetrical (i.e) it is skewed towards one side.
